```python
from bs4 import BeautifulSoup
import urllib2
import pandas as pd
```

```python
wiki = "https://en.wikipedia.org/wiki/List_of_Gaon_Digital_Chart_number_ones_of_2016"
header = {'User-Agent': 'Mozilla/5.0'} # prevent 403 error on Wikipedia
request = urllib2.Request(wiki, headers = header)
page = urllib2.urlopen(request)
soup = BeautifulSoup(page, "html.parser")
# print soup
table = soup.find("table", { "class" : "wikitable sortable" })
```

```python
# length = len(table.findAll("tr"))

df = pd.DataFrame(columns = ("week", "song", "artist", "downloads")) # index = range(length)

count = 0

for row in table.findAll("tr"):
    cells = row.findAll("td")
    # for each "tr", assign each "td" to a variable.
    if len(cells) == 5:
        entry = {"week": cells[0].find(text=True),
                "song": cells[1].find(text=True),
                "artist": cells[2].find(text=True),
                "downloads": cells[3].find(text=True)}
        df.loc[count] = entry
        count += 1
```
